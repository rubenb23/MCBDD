```python
import numpy as np
import matplotlib.pyplot as plt

# Sensitivity (fixed at 99%)
sens = 0.99

# Different specificities to compare
spec_levels = [0.99, 0.999, 0.9999, 0.99999]

# Infection prevalence range: 0.001% to 50%
prev_range = np.linspace(0.00001, 0.5, 5000)

"""
Function to calculate "P(Infected | Positive test)" or the probability someone
is actually infected if their test result is positive using Bayes' theorem
"""
def posterior_prob(prev, sens, spec):
    true_pos = sens * prev
    false_pos = (1 - spec) * (1 - prev)

    return true_pos / (true_pos + false_pos)

# Plot
plt.figure(figsize=(10, 6))

for spec in spec_levels:
    ppv = posterior_prob(prev_range, sens, spec)
    plt.plot(prev_range * 100, ppv * 100, label=f"Specificity = {spec * 100:.3f}%")

plt.xlabel("Prevalence (%)")
plt.ylabel("P(Infected | Positive test) (%)")
plt.title("Probability of actually being infected after a positive test")
plt.legend()
plt.grid(True)
plt.xlim(0, 50)
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```
